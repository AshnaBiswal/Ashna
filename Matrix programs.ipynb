```python
#Python Program to Add Two Matrices.
p=int(input("Enter the no. of rows in matrix 1 and matrix 2: "))
q=int(input("Enter the no. of cols in matrix 1 and matrix 2: "))

print("Enter elements of matrix 1: ")
m1=[[int(input())for j in range(q)]for i in range(p)]   
print("Matrix 1 :")
for m in m1:
    print(m)

print("Enter elements of matrix 2: ")
m2=[[int(input())for j in range(q)]for i in range(p)]   
print("Matrix 2 :")
for m in m2:
    print(m)
    
result=[[m1[i][j]+m2[i][j]for j in range(q)]for i in range(p)] 
print("Resultant Matrix after adding matrix 1 and matrix 2 :") 
for num in result:
    print(num)
```

```python
#Python Program to Transpose a Matrix using Nested Loop.
p=int(input("Enter the no. of rows in matrix: "))
q=int(input("Enter the no. of cols in matrix: "))

print("Enter elements of matrix : ")
m=[]
for i in range(p):
    r=[]
    for j in range(q):
        k=int(input())
        r.append(k)
    m.append(r)
print("original Matrix: ")
for i in range(p):
    for j in range(q):
        print(m[i][j]," ",end="")
    print(" ")
    
print("Matrix Transpose: ")
for i in range(p):
    for j in range(q):
        print(m[j][i]," ",end="")
    print(" ")    
```

```python
#Python Program to Transpose a Matrix using list comprehension.
p=int(input("Enter the no. of rows in matrix: "))
q=int(input("Enter the no. of cols in matrix: "))

m=[[int(input())for j in range(q)]for i in range(p)]   
print("Original Matrix :")
for i in m:
    print(i)
print("Matrix Transpose: ")
n=[[m[j][i]for j in range(q)]for i in range(p)] 
for i in n:
    print(i)
```

```python
#Python Program to Multiply Two Matrices using Nested Loop.
p=int(input("Enter the no. of rows in matrix 1: "))
n=int(input("Enter the no. of cols in matrix 1/no. of rows in matrix 2: "))
q=int(input("Enter the no. of cols in matrix 2: "))

print("Enter elements of matrix 1: ")
m1=[]
for i in range(p):
    r1=[]
    for j in range(n):
        f=int(input())
        r1.append(f)
    m1.append(r1)
print("Matrix 1 :")
for i in range(p):
    for j in range(n):
        print(m1[i][j]," ",end="")
    print(" ")

print("Enter elements of matrix 2: ")
m2=[]   
for i in range(n):
    r2=[]
    for j in range(q):
        e=int(input())
        r2.append(e)
    m2.append(r2)
print("Matrix 2 :")
for i in range(n):
    for j in range(q):
        print(m2[i][j]," ",end="")
    print(" ")
    
print("Enter",p*q,"0's:")
result=[]
for i in range(p):
    r3=[]
    for j in range(q):
        j=int(input())
        r3.append(j)
    result.append(r3)  
    
for i in range(p):
        for j in range(q):
            for k in range(n):
                result[i][j]=result[i][j]+(m1[i][k]*m2[k][j])

print("Resultant Matrix after multiplying matrix 1 and matrix 2 :") 
for i in range(p):
    for j in range(q):
        print(result[i][j]," ",end="")
    print(" ")
```

```python
#Python Program to Multiply Two Matrices using list comprehension.
p=int(input("Enter the no. of rows in matrix 1 "))
n=int(input("Enter the no. of cols in matrix 1/no. of rows in matrix 2: "))
q=int(input("Enter the no. of cols in matrix 2: "))

print("Enter elements of matrix 1: ")
m1=[[int(input())for i in range(n)]for i in range(p)]
print("Matrix 1 :")
m11=[[m1[i][j]for j in range(n)]for i in range(p)]
for i in m11:
    print(i)

print("Enter elements of matrix 2: ")    
m2=[[int(input())for i in range(q)]for i in range(n)]
print("Matrix 2 :")
m22=[[m1[i][j]for j in range(q)]for i in range(n)]

for i in m22:
    print(i)
result=[[sum(m1[i][k]*m2[k][j]for k in range(n))for j in range(q)]for i in range(p)]

print("Resultant Matrix after multiplying matrix 1 and matrix 2 :") 
for i in result:
    print(i)
```
